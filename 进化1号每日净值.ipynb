```python
import pyecharts.options as opts
from pyecharts.charts import Line
date=["22日","23日","24日","25日","26日","27日","28日","29日","31日","2日","3日","5日","6日","7日","9日"]
net_value=["1.0217","1.0101","0.9613","1.0080","1.0937","1.1066","1.1193","1.1263","1.1197","1.1174","1.1292","1.1491","1.1691","1.1741","1.1701"]
def line_base() -> Line:
    c = (
        Line()
        .add_xaxis(date)
        .add_yaxis("每日净值", net_value)
        .set_global_opts(title_opts=opts.TitleOpts(title="每日净值"))
    )
    return c
p=line_base()
```

```python
p.render_notebook()
```
